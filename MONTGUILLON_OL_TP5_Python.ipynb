**II. Traçage du graphique :**

- *Via Numpy*

````python
x = linspace(-4*pi,+4*pi,100)

y = sin(x)

plot(x,y)

legend("sin(x)")

show()
````

- *Via Matplotlib*

````python
x = np.linspace(-4*np.pi,4*np.pi,100)

y = np.cos(x)

plt.plot(x,y)

plt.legend("cos(x)")

plt.show()
````

**IX. Graphique Finale :**

- *Via Numpy*

````python
x = linspace(-4*pi,+4*pi,100)
y = sin(x)

plot(x,y,"r-",label="sin(x)",linewidth=4)

grid()

xlim(-12,12)
ylim(-1.5,1.5)

legend("sin(x)")
title("Nom de La Fonction")

xlabel("Abscisses")
ylabel("Ordonnées")



savefig("nom_de_la_figure.png")

show()
````

- *Via Matplotlib*

````python
x = np.linspace(-4*np.pi,4*np.pi,100)
y = np.cos(x)

plt.plot(x,y,"r-",label="sin(x)",linewidth=4)

plt.grid()

plt.xlim(-12,12)
plt.ylim(-1.5,1.5)

plt.legend("cos(x)")
plt.title("Nom de la Fonction")


plt.xlabel("Abscisses")
plt.ylabel("Ordonnées")



plt.show()
````

**X. Graphique Finale (+ Commentaires) :**

- *Via Matplolib*

````python
xPlot = np.linspace(-10,10) #Déclaration de l'étendue de mesure
yPlot = xPlot**2 + 3*xPlot - 4 #Déclaration de l'aspect de la courbe

plt.plot(xPlot,yPlot,"r-",label="sin(x)",linewidth=4) #Création du graphique, en fonction de l'abscicce, de l'ordonnée, on y ajoute une couleur, un label fictif, si on pointe la courbe, et une épaisseur

plt.grid() #Ajout d'une grille

plt.xlim(-10,10) #Limite l'étendue de Mesure sur l'axe des Abscisses
plt.ylim(-10,150) #Limite l'étendue de Mesure sur l'axe des Ordonnées

plt.legend("x") #On légendre notre fonction
plt.title("f(x)= x ² + 3x - 4, sur [-10,10]") #on titre notre fonction

plt.xlabel("Etendue de Mesure.") #On légende notre abscisse
plt.ylabel("Indices de X.") #On légende notre ordonnée

plt.show() #On affiche le graphique
````

- *Via Numpy*

````python
xPlot = linspace(-10,10) #Déclaration de l'étendue de mesure
yPlot = xPlot**2 + 3*xPlot - 4 #Déclaration de l'aspect de la courbe

plot(xPlot,yPlot,"r-",label="sin(x)",linewidth=4) #Création du graphique, en fonction de l'abscicce, de l'ordonnée, on y ajoute une couleur, un label fictif, si on pointe la courbe, et une épaisseur

grid() #Ajout d'une grille

xlim(-10,10) #Limite l'étendue de Mesure sur l'axe des Abscisses
ylim(-10,150) #Limite l'étendue de Mesure sur l'axe des Ordonnées

legend("x") #On légendre notre fonction
title("f(x)= x ² + 3x - 4, sur [-10,10]") #on titre notre fonction

xlabel("Etendue de Mesure.") #On légende notre abscisse
ylabel("Indices de X.") #On légende notre ordonnée

show() #On affiche le graphique
````
